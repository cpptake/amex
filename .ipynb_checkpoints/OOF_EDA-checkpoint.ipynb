# OOF EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

oof_df = pd.read_csv('oof/oof_df_v3.csv')
oof_df.head()
```

```python
sns.scatterplot(x = 'target',y = 'oof_pred',data = oof_df,)
```

```python
plt.figure(figsize=(13,10), dpi=80)
sns.violinplot(x='target', y='oof_pred', data=oof_df, scale='width', inner='quartile')
```

```python
# sns.distplot(oof_df['oof_pred'], hue = oof_df['target'],bins = 100)
g=sns.FacetGrid(oof_df, hue='target', size=5)
g.map(sns.distplot, "oof_pred",bins = 100)
g.add_legend()
```
