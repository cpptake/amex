# train testの一致度確認

```python
# train test0
train_feature = df_train.drop('customer_ID', axis = 1).columns.to_list()
test_feature0 = df_test_0.columns.to_list()

num_features0 = [col for col in train_feature if col not in test_feature0]

num_features0
```

```python
num_features0_test = [col for col in test_feature0 if col not in train_feature]

num_features0_test
```

```python
# train test1

train_feature = df_train.drop('customer_ID', axis = 1).columns.to_list()
test_feature1 = df_test_1.columns.to_list()

num_features1 = [col for col in train_feature if col not in test_feature1]

num_features1
```

```python
num_features1_test = [col for col in test_feature1 if col not in train_feature]

num_features1_test
```

```python
# train test2

train_feature = df_train.drop('customer_ID', axis = 1).columns.to_list()
test_feature2 = df_test_2.columns.to_list()

num_features2 = [col for col in train_feature if col not in test_feature2]

num_features2
```

```python
num_features2_test = [col for col in test_feature2 if col not in train_feature]

num_features2_test
```

# 結論
「num_features0_test」のリストに含まれている特徴量が被ってる特徴量のすべて  
（trainにあってtestにない特徴量はない）
「num_features0_test」をすべてDropすればいい  

```python
df_test = pd.concat([df_test_0,df_test_1,df_test_2],axis = 0).drop(num_features0_test,axis = 1)


# df_test_0 = pd.read_parquet("../feature/exp06 Basic 1500 features/test_1500features_0.parquet")
# df_test_1 = pd.read_parquet("../feature/exp06 Basic 1500 features/test_1500features_1.parquet")
# df_test_2 = pd.read_parquet("../feature/exp06 Basic 1500 features/test_1500features_2.parquet")
```
